```python
# Criar o modelo de árvore de decisão
clf = DecisionTreeClassifier()

# Treinar o modelo
clf.fit(X_train, y_train)

# Fazer previsões no conjunto de teste
y_pred = clf.predict(X_test)

# Classification Report
report1 = classification_report(y_test, y_pred)
print("Classification Report:")
print(report1)

# Calcular a accuracy do modelo
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", round(accuracy, 2))

# Calcular precisão
precision = precision_score(y_test, y_pred, average='weighted')
print("Precision:", round(precision, 2))

# Calcular recall
recall = recall_score(y_test, y_pred, average='weighted')
print("Recall:", round(recall, 2))

# Calcular F1-score
f1 = f1_score(y_test, y_pred, average='weighted')
print("F1-score:", round(f1, 2))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
classes = ['2 sets', '3 sets']

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g', xticklabels=classes, yticklabels=classes)
plt.xlabel('Pred.')
plt.ylabel('True')
plt.title('Confusion Matrix - DecisionTree')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.model_selection import StratifiedKFold
import numpy as np

# Mapear os rótulos 2 e 3 para 0 e 1, respectivamente, apenas para calcular a AUC e a curva ROC
y_train_mapped = y_train.replace({2: 0, 3: 1})
y_test_mapped = y_test.replace({2: 0, 3: 1})

# Configuração do modelo e validação cruzada, dividindo em 10 folds (k-folds)
cv = StratifiedKFold(n_splits=10)

# Criar listas para armazenar resultados
tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)

plt.figure(figsize=(8, 6))

for i, (train_index, val_index) in enumerate(cv.split(X_train, y_train_mapped)):
    # Dividir os dados
    X_fold_train, X_fold_val = X_train.iloc[train_index], X_train.iloc[val_index]
    y_fold_train, y_fold_val = y_train_mapped.iloc[train_index], y_train_mapped.iloc[val_index]

    # Treinar o modelo
    clf.fit(X_fold_train, y_fold_train)
    
    # Obter as probabilidades das previsões
    y_pred_proba = clf.predict_proba(X_fold_val)[:, 1]
    
    # Calcular a ROC e a AUC para o fold atual
    fpr, tpr, _ = roc_curve(y_fold_val, y_pred_proba)
    roc_auc = roc_auc_score(y_fold_val, y_pred_proba)
    aucs.append(roc_auc)
    plt.plot(fpr, tpr, lw=1, alpha=0.3, label=f'ROC fold {i+1} (AUC = {roc_auc:.2f})')
    tprs.append(np.interp(mean_fpr, fpr, tpr))
    tprs[-1][0] = 0.0

# Calcular a média e desvio padrão da TPR e AUC
mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = np.mean(aucs)
std_auc = np.std(aucs)

# Representar a curva ROC média
plt.plot(mean_fpr, mean_tpr, color='blue', label=f'Mean ROC (AUC = {mean_auc:.2f} ± {std_auc:.2f})', lw=2, alpha=0.8)

# Representar a linha diagonal
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - DecisionTree with Cross-Validation')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_curve, auc
import numpy as np

# Mapear os rótulos 2 e 3 para 0 e 1, respectivamente
y_train_mapped = y_train.replace({2: 0, 3: 1})
y_test_mapped = y_test.replace({2: 0, 3: 1})

# Configuração do modelo e validação cruzada, dividindo em 10 folds (k-folds)
cv = StratifiedKFold(n_splits=10)

# Inicializar listas para armazenar os resultados
tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)

plt.figure(figsize=(8, 6))

for i, (train_index, test_index) in enumerate(cv.split(X_train_important, y_train_mapped)):
    # Dividir os dados
    X_train_fold, X_test_fold = X_train_important.iloc[train_index], X_train_important.iloc[test_index]
    y_train_fold, y_test_fold = y_train_mapped.iloc[train_index], y_train_mapped.iloc[test_index]

    # Treinar o modelo
    clf_important.fit(X_train_fold, y_train_fold)

    # Calcular as probabilidades das previsões
    y_prob = clf_important.predict_proba(X_test_fold)[:, 1]

    # Calcular a curva ROC para o fold atual
    fpr, tpr, _ = roc_curve(y_test_fold, y_prob)
    tprs.append(np.interp(mean_fpr, fpr, tpr))
    tprs[-1][0] = 0.0
    roc_auc = auc(fpr, tpr)
    aucs.append(roc_auc)
    plt.plot(fpr, tpr, lw=1, alpha=0.3, label=f'ROC fold {i+1} (AUC = {roc_auc:.2f})')

mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs)

# Representar a curva de ROC média
plt.plot(mean_fpr, mean_tpr, color='blue', label=f'Mean ROC (AUC = {mean_auc:.2f} ± {std_auc:.2f})', lw=2, alpha=0.8)

# Representar a linha diagonal
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2)

# Gráfico
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - DecisionTree (Top 5 features) with Cross-Validation')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import StratifiedKFold
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, f_classif
import matplotlib.pyplot as plt
import numpy as np

# Mapear os rótulos 2 e 3, para 0 e 1 respetivamente
y_train_mapped = y_train.replace({2: 0, 3: 1})

# Configuração do modelo e validação cruzada, dividindo em 10 folds (k-folds)
cv = StratifiedKFold(n_splits=10)

# Inicializar listas para armazenar resultados
tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)

plt.figure(figsize=(8, 6))

for i, (train_index, val_index) in enumerate(cv.split(X_train_selected, y_train_mapped)):
    # Dividir os dados
    X_fold_train, X_fold_val = X_train_selected[train_index], X_train_selected[val_index]
    y_fold_train, y_fold_val = y_train_mapped.iloc[train_index], y_train_mapped.iloc[val_index]

    # Treinar o modelo
    nb_model1.fit(X_fold_train, y_fold_train)
    
    # Obter as probabilidades das previsões
    y_pred_proba = nb_model1.predict_proba(X_fold_val)[:, 1]
    
    # Calcular a ROC e a AUC para o fold atual
    fpr, tpr, _ = roc_curve(y_fold_val, y_pred_proba)
    roc_auc = auc(fpr, tpr)
    aucs.append(roc_auc)
    plt.plot(fpr, tpr, lw=1, alpha=0.3, label=f'ROC fold {i+1} (AUC = {roc_auc:.2f})')
    tprs.append(np.interp(mean_fpr, fpr, tpr))
    tprs[-1][0] = 0.0

mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = np.mean(aucs)
std_auc = np.std(aucs)

# Representar a curva de ROC média
plt.plot(mean_fpr, mean_tpr, color='blue', label=f'Mean ROC (AUC = {mean_auc:.2f} ± {std_auc:.2f})', lw=2, alpha=0.8)

# Representar a linha diagonal
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Naive Bayes (Top 5 Features) with Cross-Validation')
plt.legend(loc="lower right")
plt.show()
```

```python
# Representar as 5 melhores features
top_features_indices = selector.get_support(indices=True)
top_features_scores = selector.scores_[top_features_indices]
top_features_names = X_train.columns[top_features_indices]

# Normalizar as pontuações para que a soma seja 1
normalized_scores = top_features_scores / top_features_scores.sum()

plt.figure(figsize=(10, 6))
bars = plt.barh(range(len(normalized_scores)), normalized_scores, align='center')
plt.yticks(range(len(normalized_scores)), top_features_names)
plt.xlabel('Importância Normalizada da Feature')
plt.ylabel('Features')
plt.title('Importância das Features - K-Nearest Neighbors (Top 5 Features)')
for bar, value in zip(bars, normalized_scores):
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{value:.2f}', va='center')

plt.show()
```

# **Gradient Boosting**

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler

# Padronizar os dados
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Seleção das 5 melhores features
selector = SelectKBest(score_func=f_classif, k=5)  # Seleciona as 5 melhores características
X_train_selected = selector.fit_transform(X_train_scaled, y_train)
X_test_selected = selector.transform(X_test_scaled)

# Criar e treinar o modelo
model_gb1 = GradientBoostingClassifier()
model_gb1.fit(X_train_selected, y_train)

# Previsões
y_pred = model_gb1.predict(X_test_selected)

# Classification Report
report = classification_report(y_test, y_pred)
print("Classification Report (Grandient Boosting):")
print(report)

# Calcular accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", round(accuracy, 2))

# Calcular precision
precision = precision_score(y_test, y_pred, average='weighted')
print("Precision:", round(precision, 2))

# Calcular recall
recall = recall_score(y_test, y_pred, average='weighted')
print("Recall:", round(recall, 2))

# Calcular F1-score
f1 = f1_score(y_test, y_pred, average='weighted')
print("F1-score:", round(f1, 2))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
classes = ['2 sets', '3 sets'] 
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='g', xticklabels=classes, yticklabels=classes)
plt.xlabel('Pred.')
plt.ylabel('True')
plt.title('Confusion Matrix - Gradient Boosting')
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import numpy as np

# Mapear os rótulos 2 e 3 para 0 e 1, respetivamente
y_train_mapped = y_train.replace({2: 0, 3: 1})

# Configuração do modelo e validação cruzada, dividindo em 10 folds (k-folds)
cv = StratifiedKFold(n_splits=10)

# Criar listas para armazenar os resultados
tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)

plt.figure(figsize=(8, 6))

for i, (train_index, test_index) in enumerate(cv.split(X_train_selected, y_train_mapped)):
    X_train_fold, X_test_fold = X_train_selected[train_index], X_train_selected[test_index]
    y_train_fold, y_test_fold = y_train_mapped.iloc[train_index], y_train_mapped.iloc[test_index]

    # Treinar o modelo
    model_gb1.fit(X_train_fold, y_train_fold)

    # Calcular as probabilidades das previsões
    y_prob_gb = model_gb1.predict_proba(X_test_fold)[:, 1]

    # Calcular a curva ROC e AUC para o fold atual
    fpr, tpr, _ = roc_curve(y_test_fold, y_prob_gb)
    tprs.append(np.interp(mean_fpr, fpr, tpr))
    tprs[-1][0] = 0.0
    roc_auc = auc(fpr, tpr)
    aucs.append(roc_auc)
    plt.plot(fpr, tpr, lw=1, alpha=0.3, label=f'ROC fold {i+1} (AUC = {roc_auc:.2f})')

mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs)

# Representar a curva de ROC média
plt.plot(mean_fpr, mean_tpr, color='blue', label=f'Mean ROC (AUC = {mean_auc:.2f} ± {std_auc:.2f})', lw=2, alpha=0.8)

# Representar a linha diagonal
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2)

# Gráfico
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - GradienteBoosting (Top 5 features) with Cross-Validation')
plt.legend(loc="lower right")
plt.show()
```
